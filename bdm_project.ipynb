```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('/content/Untitled spreadsheet - Sheet1.csv')
X=dataset.iloc[:,1:7]
Y=dataset['sales']
x_train,x_test,ytrain,ytest=train_test_split(X,Y)
```

```python
X
```

```python
lin=LinearRegression()
lin.fit(x_train,ytrain)
y_pred=lin.predict(x_test)
coeff=lin.coef_
components=pd.DataFrame(zip(X.columns,coeff),columns=['components','value'])
components=components.append({'components':'intercept','value':lin.intercept_},ignore_index=True)
```
